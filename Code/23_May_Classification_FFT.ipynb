```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
from scipy.io import loadmat
```

```python
location = "C:/Users/alexc/OneDrive/Documents/Taiwan Classes/Second Semester/Brain Computer Interfaces/BCI Presentation and Project/SSVEP Projects/Wearable SSVEP BCI Database/Dataset/FFTs_from_matlab/"

files = glob.glob(location+"/*.mat")
files
```

```python
#subjects = loadmat(files[0])  #for only 1 subject at a time

subjects = []

for file in files:
    subjects.append(loadmat(file))
subjects[0]
```

```python
data = []

for i in range(len(subjects)):
    data.append(subjects[i]['fft_fir'])

data = np.array(data)
data = data.reshape(-1,60)
print(data.shape)
data_df = pd.DataFrame(data)
```

```python
data_df
```

```python
N_subjects = len(subjects)
data_reshaped = data.reshape(8,240*N_subjects,60)
channel_data = np.zeros([8,240*N_subjects], dtype=object)

for i in range(8):
    for j in range(240*N_subjects):
        channel_data[i,j] = data_reshaped[i,j,:]
```

```python
channel_data_df = pd.DataFrame(channel_data, index = ["POz", "PO3", "PO4", "PO5", "PO6", "Oz", "O1", "O2"])
final_df = channel_data_df.T
```

```python
final_df["Target"] = np.NaN

# 20X12 = 240 trials in 1 subject. To do it for several subjects just mulptiply 20 x Number of subjects (since then it will be multiply x 12 targets)
for i in range(1,21*N_subjects):
        for j in range(12):
                index = (12*(i-1))
                final_df.Target.loc[index+j] = j+1     
```
