```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
from scipy.io import loadmat
```

```python
location = "C:/Users/alexc/OneDrive/Documents/Taiwan Classes/Second Semester/Brain Computer Interfaces/BCI Presentation and Project/SSVEP Projects/Wearable SSVEP BCI Database/Dataset/FFTs_from_matlab/"

import re
 
# helper function to perform sort
def num_sort(test_string):
    return list(map(int, re.findall(r'\d+', test_string)))[0]

files = glob.glob(location+"/*.mat")
files.sort(key=num_sort)
files
```

```python
#subjects = loadmat(files[0])  #for only 1 subject at a time

subjects = []

for file in files:
    subjects.append(loadmat(file))
```

```python
subjects[1]['fft_fir']
```

```python
data = []

for i in range(len(subjects)):
    data.append(subjects[i]['fft_fir'])

data = np.array(data)
print(data.shape)

data = data.reshape(-1,60)
print(data.shape)
data_df = pd.DataFrame(data)
```

```python
data_df.to_csv("data_60_points.csv", index=False)
```

```python
N_subjects = len(subjects)
data_reshaped = data.reshape(N_subjects,8,240,60)


final_array = [] #all 102 subject

for i in range(N_subjects):
    data_subject_array = [] #all 240 trials in 1 subject
    for j in range(240):
        channel_array = [] #data in each Channel per index
        for x in range(8):
            channel_array.append(data_reshaped[i,x,j,:])
        data_subject_array.append(channel_array)
    final_array.append(data_subject_array)

final_array = np.array(final_array)
final_array.shape
```

```python
final_array[1,0,0,:]
```

```python
final_array_2d = final_array.reshape(102*240,8*60)
final_array_2d.shape
```

```python
pd.set_option('display.max_columns', None)

final_array_df = pd.DataFrame(final_array_2d)
final_array_df.iloc[240:,60:]

# final_array_df[a,b] , In "a" = each 240 belongs to a different person. B is each 60 points of each channel concatenated horizontally. 60 of channel 1 ->
# then 60 of channel 2, and so on.
```

```python
#Create Targets:

final_array_df["Target"] = np.NaN

# To add targets (1 to 12) to every row in chuncks of 12 for the length of 8 channels and 20 units (10 blocks, 2 types of electrodes), and finally
# for N_subjects (102 subjects).

for a in range(N_subjects+1):  #not sure if the +1 is needed.
        for i in range(1,21*8):
                for j in range(12):
                        index = ((12*(i-1)))
                        final_array_df.Target.loc[index+j+(1920*a)] = int(j+1)  
```

```python
final_array_df.to_csv("real_trainable_data.csv", index = False) 
```
